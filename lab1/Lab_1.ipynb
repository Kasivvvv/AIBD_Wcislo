```python
f = lambda x: x*x + 5

def plot_1(x, f_x, title):

    plt.plot(x,f_x)
    plt.legend("x")
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid()
    plt.show()

x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 =  np.linspace(0,5)

plot_1(x1,f(x1), 'Funckja dla przedziału -1,1')
plot_1(x2,f(x2), 'Funckja dla przedziału -6,6')
plot_1(x3,f(x3), 'Funckja dla przedziału 0,5')
```
